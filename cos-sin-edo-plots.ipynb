```python
# https://physics.stackexchange.com/q/637017/75633
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = False
plt.rcParams['font.family'] = 'serif'
# don't display plots in new windows:
%matplotlib inline
```

```python
#x = np.linspace(-np.pi/2,3*np.pi/2,50)
x = np.linspace(-2,5,50)
r = 0
y = -np.cos(r)*np.sin(x)
```

```python
fig1 = plt.figure()
axes = fig1.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom; width, height
# Move left y-axis and bottim x-axis to centre, passing through (0,0)
axes.spines['left'].set_position('zero')
axes.spines['bottom'].set_position('zero')
# Eliminate upper and right axes
axes.spines['right'].set_color('none')
axes.spines['top'].set_color('none')
# Show ticks in the left and lower axes only
axes.xaxis.set_ticks_position('bottom')
axes.yaxis.set_ticks_position('left')
axes.set_yticks([])
axes.set_xlabel("$x$",fontsize=16,loc='right')
axes.set_ylabel("$\dot{x}$",fontsize=16,loc='top')
axes.set_title("$r<\pi/2$")
# plot function
axes.plot(x,y)
# plot fixed points
axes.plot(0,0,color="black",markerfacecolor="black", markeredgewidth=2,marker="o",markersize=10)
axes.plot(np.pi,0,color="black",markerfacecolor="white", markeredgewidth=2,marker="o",markersize=10)
#plot arrows
matplotlib.pyplot.arrow(-1.1, 0.1, 0.5,0,head_width=0.08, head_length=0.17)
matplotlib.pyplot.arrow( 1.1, 0.1, -0.5,0,head_width=0.08, head_length=0.17)
matplotlib.pyplot.arrow( 2.7, 0.1, -0.5,0,head_width=0.08, head_length=0.17)
matplotlib.pyplot.arrow( 3.6, 0.1, 0.5,0,head_width=0.08, head_length=0.17)
fig1.savefig("/home/stafusa/stability.png")
```

```python
fig2 = plt.figure()
axes2 = fig2.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom; width, height
plt.xlim([-0.5, 3.5])
plt.ylim([-1.5, 7])
axes2.set_xlabel("$r$",fontsize=16)
axes2.set_ylabel("$x^*$",fontsize=16)
# Move left y-axis and bottim x-axis to centre, passing through (0,0)
axes2.spines['left'].set_position('zero')
axes2.spines['bottom'].set_position('zero')
# Eliminate upper and right axes
axes2.spines['right'].set_color('none')
axes2.spines['top'].set_color('none')
# Show ticks in the left and lower axes only
axes2.xaxis.set_ticks_position('bottom')
axes2.yaxis.set_ticks_position('left')
# plot
axes2.plot([0,np.pi/2],[np.pi,np.pi],color="black", linestyle='--',lw=3)
axes2.plot([np.pi/2,np.pi],[np.pi,np.pi],color="black", linestyle='-',lw=3)
axes2.plot([0,np.pi/2],[0,0],color="black", linestyle='-',lw=3)
axes2.plot([np.pi/2,np.pi],[0,0],color="black", linestyle='--',lw=3)
axes2.plot([np.pi/2,np.pi/2],[0,2*np.pi],color="black", linestyle='-',lw=3)
fig2.savefig("/home/stafusa/bifurcation.png")
```
